## Sunset Bot

```python
import requests, json, random, re, time
from IPython.display import display, Image
from bs4 import BeautifulSoup
import urllib
import twitter
```

```python
ckey = ''
csecret = ''
tkey = ''
tsecret = ''
from define import set_credential_variables
```

```python
# you can put your credentials directly here -- I just used a define.py file not to push them to GitHub
api = twitter.Api(consumer_key=ckey,
                      consumer_secret=csecret,
                      access_token_key=tkey,
                      access_token_secret=tsecret)
```

```python
search = 'sunset'
url = 'https://commons.wikimedia.org/w/api.php?action=query&list=allimages&aiprop=url|name|title|extmetadata&aifrom=' + search + '&format=json'
response = requests.get(url)
```

```python
adjectifs = ['intéressante', 'superbe', 'stupéfiante', 'stupéfiante, magique', 'impressionnante', 'incroyable', 'intrigante', '']
def get_adjective():
    # random.choice(adjectifs)
    if len(adjectifs) > 0:
        return adjectifs.pop()
    return ""

def display_image(url):
    img_response = urllib.request.urlopen(url).read()
    img_obj = Image(img_response, width=200)
    display(img_obj)

```

```python
def get_credit(extmetadata):
    if not 'Artist' in extmetadata.keys():
        return None
    author_link = extmetadata['Artist']['value']
    author_soup = BeautifulSoup(author_link, "html.parser")
    author = author_soup.find('a').string

    license = extmetadata['LicenseShortName']['value']
        
    return {'license': license, 'author': author}

```

```python
r_num = re.compile(r"(^.*)(/.{1,2}/.{1,2}/)(.*$)")
def get_thumbnail(img_link):
    match = re.split(r_num, img_link)
    if len(match) > 2:
        return match[1] + '/thumb' + match[2] + match[3] + "/800px-" + match[3]
    else:
        return None
```

```python
imgs = response.json()

r_jpg = re.compile(r"File:(.*)\.(jpg|JPG|jpeg)$")
limit = 6
i = 0
for img in imgs['query']['allimages']:
    
    if i > limit:
        break
    filename = img['title']
    title = re.match(r_jpg, filename)
    if title:
        attr = get_credit(img['extmetadata'])
        if attr:
            
            phrase = "Le wikipédien " + attr['author'] + " a nommé cette " + get_adjective() + " photo «"
            phrase += title.group(1)
            phrase += "». Elle est sous licence " + attr['license'] + ". "
            
            print(len(phrase))
            #print(img)
            print(phrase)
            
            thumb = get_thumbnail(img['url'])
            
            if thumb:
                display_image(thumb)
            else:
                print('Erreur avec', img['url'])
            
            
            result = api.PostUpdate(phrase, thumb)

            print(result)
            
            time.sleep(90)
            i += 1
        
```
